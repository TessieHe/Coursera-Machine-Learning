```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
from scipy.optimize import minimize
```

```python
plt.ion() #在Python Consol命令行中，默认是交互模式。而在python脚本中，matplotlib默认是阻塞模式。  打开交互模式

print('loading data from rating datasets....')

data = sio.loadmat('ex8_movies.mat') #dara是字典形式保存的变量

Y =data['Y']
R= data['R']

print('Y.shape:',Y.shape) #943个用户对1682部电影的评分（1-5）
print('R.shape:',R.shape) #Y（i，j）=1 当且仅当用户j对电影i有评分，否则为0
print('Y: ',Y)
print('R: ',R)
```

```python
plt.imshow(Y) 
plt.ylabel('Movies')
plt.xlabel('Users')
```

```python
def cofiCostFunc(params,Y,R,num_users,num_movies,num_features,_lambda):  #损失函数和梯度
    #电影矩阵X和用户矩阵theta都被flatten的存在params中，方便计算
    X = params[0:num_movies*num_features].reshape(num_movies,num_features)
    Theta = params[num_movies*num_features:].reshape(num_users,num_features)
    X_grad = np.zeros(X.shape)  #梯度矩阵初始化为0
    Theta_grad = np.zeros(Theta.shape)
    J = np.sum(R*np.power(np.dot(X,Theta.T)-Y ,2))/2 + \
    _lambda*np.sum(np.power(X,2))/2 + _lambda*np.sum(np.power(Theta,2))/2
    for i in range(num_movies):
        idx = np.where(R[i,:]==1)[0] #对于第i部电影有评论的用户
        tempTheta = Theta[idx,:]
        tempY = Y[i,idx]
        X_grad[i,:] = np.dot(np.dot(X[i,:],tempTheta.T)-tempY,tempTheta) + _lambda*X[i,:]  #用交互信息更新梯度
    for j in range(num_users):
        jdx = np.where(R[:,j]==1)[0] #用户j评价过的电影
        tempX = X[jdx,:]
        tempY = Y[jdx,j]
        Theta_grad[j,:] = np.dot(np.dot(tempX,Theta[j,:].T)-tempY,tempX)+ _lambda*Theta[j,:]
    return J, np.append(X_grad.flatten(), Theta_grad.flatten())  #梯度矩阵flatten成一列    

def loadMovieList():
    movielist=[]
    with open('movie_ids.txt') as f:
        lines = f.readlines()
        for line in lines: 
            if line=='\n':
                pass
            else:movielist.append(line[line.find(' ')+1:])
    return movielist 
```

```python
movieList = loadMovieList()
n_movies=len(movieList)
```

# 增加新用户

```python
#来一个新用户时，将新用户的评分加到矩阵Y中，同时改变R
#  Initialize my ratings
my_ratings = np.zeros(n_movies)

# Check the file movie_idx.txt for id of each movie in our dataset
# For example, Toy Story (1995) has ID 1, so to rate it "4", you can set
my_ratings[0] = 4

# Or suppose did not enjoy Silence of the Lambs (1991), you can set
my_ratings[97] = 2

# We have selected a few movies we liked / did not like and the ratings we
# gave are as follows:
my_ratings[6] = 3
my_ratings[11]= 5
my_ratings[53] = 4
my_ratings[63]= 5
my_ratings[65]= 3
my_ratings[68] = 5
my_ratings[182] = 4
my_ratings[225] = 5
my_ratings[354]= 5


Y = np.vstack((my_ratings,Y.T)).T #增加一列在最前面

Temp = np.zeros(R.shape[0])
Temp[np.where(my_ratings !=0 )] =1 #新增一列R
R = np.vstack((Temp,R.T)).T
```

# 归一化

```python
# 归一化分数矩阵Y
def normalizeRatings(Y,R):
    n_movies,n_users = Y.shape
    Ymean = np.zeros(n_movies)  #shape[n_movies,]
    Ynorm = np.zeros(Y.shape)  #初始化为0矩阵
    
    for i in range(n_movies):
        idx = np.where(R[i,:]==1)[0]
        Ymean[i] = np.mean(Y[i,idx])  #所有用户对第i部电影平均评分
        Ynorm[i,idx] = Ynorm[i,idx]-Ymean[i]
        
    return  Ynorm,Ymean
```

```python
Ynorm,Ymean = normalizeRatings(Y,R)

num_users = Ynorm.shape[1]
num_movies = Ynorm.shape[0]
num_features = 10 # 自己定义
```

# 训练X theta

```python
# 初始化
X = np.random.randn(num_movies,num_features)
Theta = np.random.randn(num_users,num_features)
initial_parameters = np.append(X.flatten(),Theta.flatten())
_lambda = 10
```

## 方法一：scipy.optimize.minimize

```python
#迭代
costFunction = lambda t: cofiCostFunc(t,Y,R,num_users,num_movies,num_features,_lambda)
res = minimize(costFunction,initial_parameters,method='CG',jac=True,options={'maxiter':100})
print(res)
```

## 方法二：手写GC back propagation

```python
def iteration(iter_time,X_in,Theta_in,Y,R,num_users,num_movies,num_features,_lambda,alpha):
    params = np.append(X_in.flatten(),Theta_in.flatten())
    cost=[]
    for i in range(iter_time):
        J,params_grad = cofiCostFunc(params,Y,R,num_users,num_movies,num_features,_lambda)
        if i%10==0:
            print('cost after ',i,' :', J)
        cost.append(J)
        params = params - alpha*params_grad
    plt.plot(cost)
    return params
        
```

```python
X_in = np.random.randn(num_movies,num_features)
Theta_in = np.random.randn(num_users,num_features)
params_GC=iteration(100,X_in,Theta_in,Y,R,num_users,num_movies,num_features,_lambda,0.001)
```

```python
X_GC = params_GC[0:num_movies*num_features].reshape(num_movies,num_features)  #抽取X，Theta
Theta_GC = params_GC[num_movies*num_features:].reshape(num_users,num_features)

Y_hat_GC = np.dot(X_GC,Theta_GC.T) #计算新的评分矩阵

my_predictions_GC = Y_hat_GC[:,0]+ Ymean #归一化

idx_GC = np.argsort(-my_predictions_GC)

print('top 5 for u:')
for i in range(5):
    movie = movieList[idx_GC[i]]
    print('movie:',movie)
```
